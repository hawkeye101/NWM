# DATA PREPROCESSING
1) Correcting 'AC name' spellings

```python
for i in range(1, np.shape(df['AC No'].unique())[0] + 1):
    AC = df.loc[df['AC No'] == i, 'AC name'].unique()
    if AC.size == 2 and 'SC' not in AC[1] and 'ST' not in AC[1]:
        df['AC name'] = df['AC name'].replace([ AC[1] ], AC[0])
```

```python
df['AC name'].unique().size
```

• Size reduced from 289 to 281

```python
for i in range(1, np.shape(df['AC No'].unique())[0] + 1):
    a = df.loc[df['AC No'] == i, 'AC name'].unique()
    print(a)
```

2) All party names are made lowercase because there is discrepancy in data.

```python
df['Party'] = df['Party'].str.lower()
```

```python
df['Party']
```

3) Further corrections based on observed discrepancies.

```python
df['Party'] = df['Party'].replace(['akhil bharitya mithila party'],'akhil bhartiya mithila party')
df['Party'] = df['Party'].replace(['proutist bloc india'],'proutist bloc of india')
df['Party'] = df['Party'].replace(['proutist sarva samaj party'],'proutist sarva samaj')
df['Party'] = df['Party'].replace(['janvadi party(socialist)'],'janvadi party (socialist)')
```

• total_votes tells you total votes for each party in each election.

```python
yunique = df['Election Year'].unique()
punique = df['Party'].unique()
len1 = len(yunique)
len2 = len(punique)
total_votes = np.zeros((len1, len2))
for index1 in range(len1):
    #print(index1)
    #print(df['Election Year'][index1])
    for index2 in range(len2):
        total_votes[index1, index2] = df.loc[(df['Election Year'] == yunique[index1]) & (df['Party'] == punique[index2]), 'Votes'].sum()
```

```python
yunique
```

• total_votes_seats tells you total votes for each party in each election in each constituency.

```python
cunique = df['AC name'].unique()
len3 = len(cunique)
total_votes_seats = np.zeros((len1, len2, len3))
for index1 in range(len1):
    for index2 in range(len2):
        #print(index2)
        for index3 in range(len3):
            total_votes_seats[index1, index2, index3] = df.loc[(df['Election Year'] == yunique[index1]) & (df['Party'] == punique[index2]) & (df['AC name'] == cunique[index3]), 'Votes'].sum()
```
